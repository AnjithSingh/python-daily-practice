Check the number is Strong number
A Strong number is a special number whose sum of the all-digit factorial should be equal to thenumber itself.
For example - The given number is 145, we have to pick each digit and find the factorial 1! = 1, 4! =24, and 5! = 120.

```python
def fact(num):
    fact = 1
    for i in range(1, num + 1):
        fact = fact * i
    return fact

num = int(input("Enter the number: "))
summ = 0
num1 = num

while num != 0:
    rem = num % 10
    num = num // 10
    summ = summ + fact(rem)

if summ == num1:
    print("Strong Number")
else:
    print("Not Strong Number")
```
