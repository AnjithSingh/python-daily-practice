# Series


```python
import pandas as pd
```

```python
x = [3,434,53,42]

var = pd.Series(x, index = ['a', 'b' , 'c' , 'd'] , dtype = 'float', name = 'Python')
print(var)
print(type(var))
print(var[2])
```

```python
dictionary = {
    'name' : ['python' , 'java ', 'c' , 'html'],
    'rank' : [1 , 2 , 3 ,4 ],
    'power' : [100 , 200, 100 ,9999]
}

var1 = pd.Series(dictionary)
print(var1)
```

# DataFrame

```python
import pandas as pd
l = [1,2,3,4,5,6]

var = pd.DataFrame(l)

print(type(var))
print(var)
```

```python
var = pd.DataFrame({'A': [1,2,3,4,5] , 'B' : [6,7,8,9,10]})

print(var)
```

```python
var['C'] = var['A'] + var['B']

print(var)
```

```python
var['D'] = var['A'] - var['B']

print(var)
```

```python
var['E'] = var['A'] / var['B']

print(var)
```

```python
var['F'] = var['A'] * var['B']

print(var)
```

```python
import pandas as pd
import numpy as np
import random

# Set a seed so the random numbers are the same every time
np.random.seed(42)

num_rows = 1000

data = {
    'Employee_ID': range(1, num_rows + 1),
    'Department': np.random.choice(['HR', 'Engineering', 'Sales', 'Marketing'], num_rows),
    'Age': np.random.randint(22, 60, num_rows),
    'Salary': np.random.randint(40000, 150000, num_rows),
    'Years_Experience': np.random.randint(0, 20, num_rows),
    'Performance_Score': np.random.uniform(1.0, 5.0, num_rows).round(1)
}

df = pd.DataFrame(data)

df.to_csv('large_employee_data.csv', index=False)
print(f"File 'large_employee_data.csv' created with {num_rows} rows.")
```
