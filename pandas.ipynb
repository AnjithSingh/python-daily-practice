#Series


```python
import pandas as pd
```

```python
x = [3,434,53,42]

var = pd.Series(x, index = ['a', 'b' , 'c' , 'd'] , dtype = 'float', name = 'Python')
print(var)
print(type(var))
print(var[2])
```

```python
dictionary = {
    'name' : ['python' , 'java ', 'c' , 'html'],
    'rank' : [1 , 2 , 3 ,4 ],
    'power' : [100 , 200, 100 ,9999]
}

var1 = pd.Series(dictionary)
print(var1)
```

```python
s = pd.Series(12 , index = [1,2,3,4,5])
print(s)
```

```python
s1 = pd.Series(12 , index = [1,2,3,4,5,6,7,8,9])
s2 = pd.Series(12 , index = [1,2,3,4,5])
print(s1+s2)
```
