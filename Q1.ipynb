 
ABC is a right triangle,  90 at B .
Therefore, angle ABC = 90 .

Point M is the midpoint of hypotenuse .

You are given the lengths AB and BC.
Your task is to find Angle MBC (angle theta , as shown in the figure) in degrees.

```python
import math
AB = int(input())
BC = int(input())

# 2. Calculate the angle in radians
# We know that angle MBC = angle ACB.
# tan(angle) = AB / BC, so angle = arctan(AB / BC)
theta_radians = math.atan(AB / BC)

# 3. Convert radians to degrees
theta_degrees = math.degrees(theta_radians)

# 4. Handle Rounding (Round half up)
# Python 3's built-in round() rounds to the nearest EVEN number for .5 cases (e.g., 56.5 -> 56).
# The problem requires 56.5 -> 57. 
# Adding 0.5 and casting to int achieves the "round half up" behavior for positive numbers.
rounded_angle = int(theta_degrees + 0.5)

# 5. Print the output with the degree symbol
print(f"{rounded_angle}{chr(176)}")
```
